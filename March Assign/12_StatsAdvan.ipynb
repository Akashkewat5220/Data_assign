```python
import numpy as np
import scipy.stats as stats

observed = np.array([10, 20, 30, 15, 12, 13])  # Observed frequencies
expected = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed)  # Expected frequencies

chi2, p_value = stats.chisquare(observed, f_exp=expected)

print("Chi-square statistic:", chi2)
print("P-value:", p_value)
```

```python
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 50

confidence_interval = stats.t.interval(0.9, df=sample_size - 1, loc=sample_mean, scale=sample_std / (sample_size ** 0.5))

print("90% Confidence Interval:", confidence_interval)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

df = 10  # Degrees of freedom
x = np.linspace(0, 30, 500)  # Range of chi-square values

# Plot the chi-square distribution
plt.plot(x, stats.chi2.pdf(x, df), label='chi-square distribution (df=10)')
plt.xlabel('Chi-square value')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution')
plt.legend()

# Shade the area corresponding to chi-square statistic of 15
x_fill = np.linspace(15, 30, 100)
plt.fill_between(x_fill, stats.chi2.pdf(x_fill, df), color='gray', alpha=0.5)

plt.show()
```

```python
import numpy as np
import scipy.stats as stats

observed = np.array([45, 55])  # Observed frequencies (tails, heads)
expected = np.array([0.5, 0.5]) * np.sum(observed)  # Expected frequencies (fair coin)

chi2, p_value = stats.chisquare(observed, f_exp=expected)

print("Chi-square statistic:", chi2)
print("P-value:", p_value)
```

```python
import scipy.stats as stats

sample_mean = 72
sample_std = 10
sample_size = 30
population_mean = 70

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std / (sample_size ** 0.5))

# Calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=sample_size - 1))

print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
